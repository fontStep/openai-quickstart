```python
# 定义要处理的文本
text = """
这个数据集包含两个测试样本，由OpenAI的AI语言模型ChatGPT提供。
这些样本包括一个Markdown表格和一个英文文本段落，可以用来测试支持文本和表格格式的英文到中文翻译软件。
文本测试
敏捷的棕色狐狸跳过懒惰的狗。这个句子包含了英语字母表中的每一个字母。句子通常被用来测试字体、键盘和其他文本相关的工具。除了英语，很多其他语言也有句子。一些句子由于语言 的独特特征更难编写。
表格测试
"""

# 将文本按行分割成列表
lines = text.split("\n")

print("原始文本：",lines)

# 在每行末尾增加换行符
formatted_text = "\n".join([line + "\n" for line in lines])

# 输出处理后的文本
print(formatted_text)
```
